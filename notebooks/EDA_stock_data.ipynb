# Exploratory Data Analysis - Yahoo finance

```python
from datetime import date
import yfinance as yf
from plotly import graph_objs as go
```

### Constants

```python
START = "2015-01-01"
TODAY = date.today().strftime("%Y-%m-%d")
stocks = ('GOOG', 'AAPL', 'MSFT', 'GME', '1810.HK')
```

#### Download Data

```python
data = yf.download(stocks[4], START, TODAY)
#data.reset_index(inplace=True)
```

```python
data.head()
```

#### Plot Data

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=data.index, y=data['Open'], name="stock_open"))
fig.add_trace(go.Scatter(x=data.index, y=data['Close'], name="stock_close"))
fig.update_layout(
    title='Time Series Data with Rangeslider',
    xaxis_title='Date',
    yaxis_title='Price',
    xaxis_rangeslider_visible=True
)
fig.show()
```
